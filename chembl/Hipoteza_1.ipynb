```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

model = KMeans(n_clusters=3)
model.fit(df)

cluster_labels = model.predict(df)
print(cluster_labels)

df['Cluster'] = cluster_labels

cluster_counts = df['Cluster'].value_counts()

cluster_names = {0: '1', 1: '2', 2:'3'} #blokowanie parametru, żeby zaczynał od konkretnego klastra?

cluster_counts.index = [cluster_names.get(x) for x in cluster_counts.index]

plt.bar(cluster_counts.index, cluster_counts.values, color=['pink', 'violet', 'black'])
plt.xlabel('Klaster')
plt.ylabel('Liczba targetów')
plt.title('Liczba targetów w klastrach KMeans z dwoma klastrami')
plt.xticks(rotation=0) 
plt.show()
```

```python
km = KMeans (n_clusters = 3)
km
```

```python
y_predicted = km.fit_predict(df[['MolecularWeight','XLogP','HBondDonorCount','HBondAcceptorCount']])
y_predicted
```

```python
df1 = df[df.Cluster==0]
df2 = df[df.Cluster==1]
df3 = df[df.Cluster==2]

plt.scatter(df1.MolecularWeight,df1['XLogP'],color='pink')
plt.scatter(df2.MolecularWeight,df2['XLogP'],color='violet')
plt.scatter(df3.MolecularWeight,df3['XLogP'],color='black')


plt.xlabel('MolecularWeight')
plt.ylabel('XLogP')
plt.legend()
```
